```python
from scipy.stats import shapiro
import numpy as np
import pandas as pd
import scipy.stats as stats
import pylab
```

```python
data = np.random.normal(loc=20, scale=5, size=1000)
```

```python
shapiro(data)
```

```python
df = pd.DataFrame(data)
df.hist(bins=100)
```

```python
stats.probplot(data , dist='norm', plot=pylab)
```
